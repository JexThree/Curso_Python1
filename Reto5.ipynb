## Ejercicio 1: Funciones


De nuestro código que evaluaba cual era una el número mayor, convertirla a función.

Entrada: num1, num2

Salida: 

1. El num1 es mayor que num2
2. El num2 es mayor que num2
3. Ambos son iguales

````python
def numero_mayor (numero1, numero2):
    if numero1 < numero2:
        numeromayor = numero2
        print (f'El número mayor es: {numeromayor}')
    elif numero2 < numero1:
        numeromayor = numero1
        print (f'El número mayor es: {numeromayor}')
    else:
        print (f'Ambos números son iguales.')
````

````python
X = int(input("Introduce un número: "))
Y = int(input("Introduce un número: "))
numero_mayor (X, Y)
````

## Extra points

Hacer una función que nos imprima cuantos números pares e impares tiene una lista.

Para evaluar un par o impar puedes utilizar el

```Python
if numero%2 = 0:
    print("numero es par")
else:
    print("numero es impar")
```

|Entrada | Salida |
| - | - | 
| [2,3,6,8,9] | Tiene `3` pares y `2` inpares |

````python
def par_inpar (lista = []):
    for e in lista:
        if e%2 == 0:
            print('Numero Par')
        else:
            print ('Numero Inpar')
````

````python
par_inpar(1, 2, 3, 4, 5, 6)
````
